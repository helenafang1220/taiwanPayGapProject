```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
regular_income = pd.read_csv("data/Regular income.csv")
```

#### Regular Salary Comparison 

```python
fig, ax = plt.subplots(figsize=(10,5))

palette = ['#F97990','#FF9F44','#F663EC','#2FC490','#14B8CD']

sns.lineplot(x="Year", y="Regular_Salary",
             hue="Industry", palette=palette,
             data=regular_income)

sns.set_theme(context="notebook",style='white', font_scale=1.2)

plt.title(label="Regular Monthly Salary Comparisons Amongst Industries",
         fontdict={'fontsize':14})
ax.set_ylabel('Salary($NTD/Month)',fontname="helvetica neue")

plt.xticks(np.arange(0, 21, 5))
plt.legend(bbox_to_anchor=(1,1))
plt.tight_layout()
plt.savefig('img/regular_income_industry.jpg', dpi=150)
```

```python
quango = gov_em["Percentage"][gov_em['Category']=="Quango Sector"]
central = gov_em["Percentage"][gov_em['Category']=="Central Government"]
local = gov_em["Percentage"][gov_em['Category']=="Local Government"]
```

```python
quango
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Male', 'Female'
sizes = quango
size1 = central
size2 = local

colors = ['#3ba3ec','#f77189']

#seaborn setting
sns.set(font_scale=3)

fig = plt.figure(figsize=(20,10))
# fig.suptitle("2020 Government Employment From Genders")

ax1 = fig.add_subplot(131)
ax1.set_title('Quango Sector')
ax1.pie(sizes, autopct='%1.1f%%', colors = colors)

ax2 = fig.add_subplot(132)
ax2.set_title('Central Government')
ax2.pie(size1, autopct='%1.1f%%', colors = colors)

ax3 = fig.add_subplot(133)
ax3.set_title('Local Government')
ax3.pie(size2, autopct='%1.1f%%', colors = colors)

ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.legend(labels=labels, bbox_to_anchor=(1,1))
plt.tight_layout()
plt.savefig('img/2020_gov_employment.jpg', dpi=150)
plt.show()
```

```python
import textwrap
def wrap_labels(ax, width, break_long_words=False):
    labels = []
    for label in ax.get_xticklabels():
        text = label.get_text()
        labels.append(textwrap.fill(text, width=width,
                      break_long_words=break_long_words))
    ax.set_xticklabels(labels, rotation=0)
```
